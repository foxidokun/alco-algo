```python
import csv

x = []
y = []

with open('binary_heap_results.csv') as csvfile:
    sortreader = csv.reader(csvfile, delimiter=',')

    for row in sortreader:
        x.append(int(row[1]))
        y.append(int(row[2]))
```

## Binary Heap

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel('Element count', color = 'gray')
plt.ylabel('Time (us)', color = 'gray')

plt.title('Binary Heap', fontsize = 15)

plt.scatter(x, y)

plt.legend(['Binary'])
plt.show()
```

### K Heap

```python
import csv

sorts = {
    "K Heap <2>" : ([], []),
    "K Heap <4>" : ([], []),
    "K Heap <8>" : ([], []),
    "K Heap <16>" : ([], []),
    "K Heap <32>" : ([], []),
    "K Heap <64>" : ([], []),
    "K Heap <128>" : ([], []),
    "K Heap <256>" : ([], []),
    "K Heap <512>" : ([], []),
    "K Heap <1024>" : ([], []),
    "K Heap <2048>" : ([], []),
}

with open('norm_k_results2.csv') as csvfile:
    sortreader = csv.reader(csvfile, delimiter=',')
    
    for row in sortreader:
        sorts[row[0]][0].append(int(row[1]))
        sorts[row[0]][1].append(int(row[2])*100)
```

```python
plt.xlabel('Element count', color = 'gray')
plt.ylabel('Time', color = 'gray')

plt.title('K Heaps', fontsize = 15)

plt.scatter(sorts["K Heap <2>"][0], sorts["K Heap <2>"][1], s = 3)
plt.scatter(sorts["K Heap <4>"][0], sorts["K Heap <4>"][1], s = 3)
plt.scatter(sorts["K Heap <8>"][0], sorts["K Heap <8>"][1], s = 3)
plt.scatter(sorts["K Heap <16>"][0], sorts["K Heap <16>"][1], s = 3)
plt.scatter(sorts["K Heap <32>"][0], sorts["K Heap <32>"][1], s = 3)
plt.scatter(sorts["K Heap <64>"][0], sorts["K Heap <64>"][1], s = 3)
plt.scatter(sorts["K Heap <128>"][0], sorts["K Heap <128>"][1], s = 3)
plt.scatter(sorts["K Heap <256>"][0], sorts["K Heap <256>"][1], s = 3)
plt.scatter(sorts["K Heap <512>"][0], sorts["K Heap <512>"][1], s = 3)
plt.scatter(sorts["K Heap <1024>"][0], sorts["K Heap <1024>"][1], s = 3)
plt.scatter(sorts["K Heap <2048>"][0], sorts["K Heap <2048>"][1], s = 3)

plt.legend(['K Heap <2>', 'K Heap <4>', 'K Heap <8>', 'K Heap <16>', 'K Heap <32>', 'K Heap <64>', 'K Heap <128>', 'K Heap <256>', 'K Heap <512>', 'K Heap <1024>', 'K Heap <2048>'])
plt.show()
```

Сузим на большие N

```python
plt.xlabel('Element count', color = 'gray')
plt.ylabel('Time', color = 'gray')

plt.title('K Heaps', fontsize = 15)

OFFSET = 15

plt.scatter(sorts["K Heap <2>"][0][OFFSET:], sorts["K Heap <2>"][1][OFFSET:], s = 3)
plt.scatter(sorts["K Heap <4>"][0][OFFSET:], sorts["K Heap <4>"][1][OFFSET:], s = 3)
plt.scatter(sorts["K Heap <8>"][0][OFFSET:], sorts["K Heap <8>"][1][OFFSET:], s = 3)
plt.scatter(sorts["K Heap <16>"][0][OFFSET:], sorts["K Heap <16>"][1][OFFSET:], s = 3)
plt.scatter(sorts["K Heap <32>"][0][OFFSET:], sorts["K Heap <32>"][1][OFFSET:], s = 3)
plt.scatter(sorts["K Heap <64>"][0][OFFSET:], sorts["K Heap <64>"][1][OFFSET:], s = 3)
plt.scatter(sorts["K Heap <128>"][0][OFFSET:], sorts["K Heap <128>"][1][OFFSET:], s = 3)
plt.scatter(sorts["K Heap <256>"][0][OFFSET:], sorts["K Heap <256>"][1][OFFSET:], s = 3)
plt.scatter(sorts["K Heap <512>"][0][OFFSET:], sorts["K Heap <512>"][1][OFFSET:], s = 3)
plt.scatter(sorts["K Heap <1024>"][0][OFFSET:], sorts["K Heap <1024>"][1][OFFSET:], s = 3)
plt.scatter(sorts["K Heap <2048>"][0][OFFSET:], sorts["K Heap <2048>"][1][OFFSET:], s = 3)

plt.legend(['K Heap <2>', 'K Heap <4>', 'K Heap <8>', 'K Heap <16>', 'K Heap <32>', 'K Heap <64>', 'K Heap <128>', 'K Heap <256>', 'K Heap <512>', 'K Heap <1024>', 'K Heap <2048>'])
plt.show()
```
