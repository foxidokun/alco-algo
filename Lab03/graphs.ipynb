```python
import csv
from math import log

sorts = {
    "Bubble Sort" : ([], []),
    "Selection Sort" : ([], []),
    "Quick Median Sort" : ([], []),
    "Quick Central Sort" : ([], []),
    "Quick Random Sort" : ([], []),
    "Radix Sort" : ([], []),
    "Insertion Sort Op" : ([], []),
    "Insertion Sort Unop" : ([], []),
    "Insertion Sort BSearch" : ([], []),
    "Merge Sort #0"  : ([], []),
    "Merge Sort #4"  : ([], []),
    "Merge Sort #8"  : ([], []),
    "Merge Sort #16" : ([], []),
    "Merge Sort #32" : ([], []),
    "Merge Sort #64" : ([], []),
    "Merge Sort #128" : ([], []),
    "Merge Sort #256" : ([], []),
    "Merge Sort #512" : ([], []),
}

with open('sorts.txt') as csvfile:
    sortreader = csv.reader(csvfile, delimiter=',')
    
    for row in sortreader:
        sorts[row[0]][0].append(int(row[1]))
        sorts[row[0]][1].append(int(row[2]))
```

## Квадратичные сортировки

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel('Element count', color = 'gray')
plt.ylabel('Time (us)', color = 'gray')

plt.title('Quadratic sortings comparison ', fontsize = 15)

plt.scatter(sorts['Bubble Sort'][0], sorts['Bubble Sort'][1])
plt.scatter(sorts['Selection Sort'][0], sorts['Selection Sort'][1])
plt.scatter(sorts['Insertion Sort Op'][0], sorts['Insertion Sort Op'][1])
plt.scatter(sorts['Insertion Sort Unop'][0], sorts['Insertion Sort Unop'][1])
plt.scatter(sorts['Insertion Sort BSearch'][0], sorts['Insertion Sort BSearch'][1])

plt.legend(['Bubble', 'Selection', 'Insertion Optimised', 'Insertion Unoptimised', 'Insertion + Binary Search'])
plt.show()
```

**Сузим на меньшую область**

```python
smaller_range_sorts = {
    "Bubble Sort" : ([], []),
    "Selection Sort" : ([], []),
    "Insertion Sort Op" : ([], []),
    "Insertion Sort Unop" : ([], []),
    "Insertion Sort BSearch" : ([], []),
}

for name in smaller_range_sorts:
    for (x, y) in zip(sorts[name][0], sorts[name][1]):
        if (x < 200000):
            smaller_range_sorts[name][0].append(x)
            smaller_range_sorts[name][1].append(y)

            
plt.xlabel('Element count', color = 'gray')
plt.ylabel('Time (us)', color = 'gray')

plt.title('Quadratic sortings comparison ', fontsize = 15)

plt.scatter(smaller_range_sorts['Bubble Sort'][0], smaller_range_sorts['Bubble Sort'][1])
plt.scatter(smaller_range_sorts['Selection Sort'][0], smaller_range_sorts['Selection Sort'][1])
plt.scatter(smaller_range_sorts['Insertion Sort Op'][0], smaller_range_sorts['Insertion Sort Op'][1])
plt.scatter(smaller_range_sorts['Insertion Sort Unop'][0], smaller_range_sorts['Insertion Sort Unop'][1])
plt.scatter(smaller_range_sorts['Insertion Sort BSearch'][0], smaller_range_sorts['Insertion Sort BSearch'][1])

plt.legend(['Bubble', 'Selection', 'Insertion Optimised', 'Insertion Unoptimised', 'Insertion + Binary Search'])
plt.show()
```

**Сравним реализации Insertion Sort**

```python
plt.xlabel('Element count', color = 'gray')
plt.ylabel('Time (us)', color = 'gray')

plt.title('Insertion sortings comparison ', fontsize = 15)

plt.scatter(smaller_range_sorts['Insertion Sort Op'][0], smaller_range_sorts['Insertion Sort Op'][1])
plt.scatter(smaller_range_sorts['Insertion Sort Unop'][0], smaller_range_sorts['Insertion Sort Unop'][1])
plt.scatter(smaller_range_sorts['Insertion Sort BSearch'][0], smaller_range_sorts['Insertion Sort BSearch'][1])

plt.legend(['Insertion Optimised', 'Insertion Unoptimised', 'Insertion + Binary Search'])
plt.show()
```

## Выводы

1. O(nlong) работает быстрее, чем O(n^2) при достаточно больших n. Не могу не отметить свое изумление тем фактом, что пузырьковая сортировка так сильно проигрывает остальным квадратичным.

2. Insertion sort с оптимизацией (перезаписывание, а не обмены) работает быстрее обычного так как совершает меньше действий и устраняет зависимости по данным, позволяя процессору эффективнее использовать конвейер. Еще сильнее эта идея развивается в Insertion sort с бинпоиском, где убрано чтение из массива при проверке условия цикла.

3. Оптимальный размер для оптимизации -- 64 элемента. Далее квадратичная природа Insertion Sort перевешивает рекурсивную природу Merge Sort.

4. Самая лучшая сортировка Radix Sort, так как она использует знание об элементах (что их можно поразрядно сравнивать)

5. qsort на случайном масиве работает во всех реализациях примерно одинаково, выделяется только реализаци со случайным опорным, так как она требует дорогого вызова rand()
