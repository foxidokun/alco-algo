```python
import csv
from math import log

bubble_sort_x = []
bubble_sort_logx = []
bubble_sort_y = []
bubble_sort_logy = []
selection_sort_x = []
selection_sort_logx = []
selection_sort_y = []
selection_sort_logy = []
insertion_sort_x = []
insertion_sort_logx = []
insertion_sort_y = []
insertion_sort_logy = []

qsort_median_sort_x = []
qsort_median_sort_y = []
qsort_random_sort_x = []
qsort_random_sort_y = []
qsort_central_sort_x = []
qsort_central_sort_y = []

merge_sort_x = []
merge_sort_y = []
radix_sort_x = []
radix_sort_y = []


with open('sorts.txt') as csvfile:
    sortreader = csv.reader(csvfile, delimiter=',')
    
    for row in sortreader:
        if (row[0] == "Bubble Sort"):
            bubble_sort_x.append((int(row[1])))
            bubble_sort_logx.append(log(int(row[1])))
            bubble_sort_y.append((int(row[2])))
            bubble_sort_logy.append(log(int(row[2])))
        elif (row[0] == "Selection Sort"):
            selection_sort_x.append((int(row[1])))
            selection_sort_logx.append(log(int(row[1])))
            selection_sort_y.append((int(row[2])))
            selection_sort_logy.append(log(int(row[2])))
        elif (row[0] == "Insertion Sort"):
            insertion_sort_x.append((int(row[1])))
            insertion_sort_logx.append(log(int(row[1])))
            insertion_sort_y.append((int(row[2])))
            insertion_sort_logy.append(log(int(row[2])))
        elif (row[0] == "Quick Median Sort"):
            qsort_median_sort_x.append((int(row[1])))
            qsort_median_sort_y.append((int(row[2])))
        elif (row[0] == "Quick Random Sort"):
            qsort_random_sort_x.append((int(row[1])))
            qsort_random_sort_y.append((int(row[2])))
        elif (row[0] == "Quick Central Sort"):
            qsort_central_sort_x.append((int(row[1])))
            qsort_central_sort_y.append((int(row[2])))         
        elif (row[0] == "Merge Sort"):
            merge_sort_x.append((int(row[1])))
            merge_sort_y.append((int(row[2])))
        elif (row[0] == "Radix Sort"):
            radix_sort_x.append((int(row[1])))
            radix_sort_y.append((int(row[2])))
```

## Квадратичные сортировки

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel('Element count', color = 'gray')
plt.ylabel('Time (us)', color = 'gray')

plt.title('Quadratic sortings comparison ', fontsize = 15)

plt.scatter(bubble_sort_x, bubble_sort_y)
plt.scatter(selection_sort_x, selection_sort_y)
plt.scatter(insertion_sort_x, insertion_sort_y)

plt.legend(['Bubble', 'Selection', 'Insertion'])
plt.show()
```

```python
plt.xlabel('ln(Element count)', color = 'gray')
plt.ylabel('ln(Time (us))', color = 'gray')

plt.title('Quadratic sortings comparison ', fontsize = 15)

plt.scatter(bubble_sort_logx, bubble_sort_logy)
plt.scatter(selection_sort_logx, selection_sort_logy)
plt.scatter(insertion_sort_logx, insertion_sort_logy)

plt.legend(['Bubble', 'Selection', 'Insertion'])
plt.show()
```

## QSORT

```python
plt.xlabel('Element count', color = 'gray')
plt.ylabel('Time (us)', color = 'gray')

plt.title('Qsort sortings comparison ', fontsize = 15)

plt.scatter(qsort_median_sort_x, qsort_median_sort_y, s = 3)
plt.scatter(qsort_central_sort_x, qsort_central_sort_y, s = 3)
plt.scatter(qsort_random_sort_x, qsort_random_sort_y, s = 3)

plt.legend(['Median', 'Central', 'Random'])
plt.show()
```

## Выводы

O(nlong) работает быстрее, чем O(n^2) при достаточно больших n. Не могу не отметить свое изумление тем фактом, что пузырьковая сортировка так сильно проигрывает остальным квадратичным.


Можно также отметить тот факт, что quicksort со случайным выбором опорного элемента не выигрывает у остальных реализаций из-за характера массива: так как он из случайных чисел, то выбор медианы или среднего равносильны выбору случайного, однако требуют меньше действий (ведь rand это дорогой вызов функции), а потому быстрее.
