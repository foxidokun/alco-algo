```python
import csv
from math import log

sorts = {
    "Bubble Sort" : ([], []),
    "Selection Sort" : ([], []),
    "Quick Median Sort" : ([], []),
    "Quick Central Sort" : ([], []),
    "Quick Random Sort" : ([], []),
    
    "Quick Median Sort :: EQUAL" : ([], []),
    "Quick Central Sort :: EQUAL" : ([], []),
    "Quick Random Sort :: EQUAL" : ([], []),

    "Quick Median Sort :: ZEBRA" : ([], []),
    "Quick Central Sort :: ZEBRA" : ([], []),
    "Quick Random Sort :: ZEBRA" : ([], []),

    "Quick Median Sort :: LOTSAME" : ([], []),
    "Quick Central Sort :: LOTSAME" : ([], []),
    "Quick Random Sort :: LOTSAME" : ([], []),
    
    "Quick Median Sort :: USHAPE" : ([], []),
    "Quick Central Sort :: USHAPE" : ([], []),
    "Quick Random Sort :: USHAPE" : ([], []),


    "Radix Sort" : ([], []),
    "Insertion Sort Op" : ([], []),
    "Insertion Sort Unop" : ([], []),
    "Insertion Sort BSearch" : ([], []),
    "Merge Sort #0"  : ([], []),
    "Merge Sort #4"  : ([], []),
    "Merge Sort #8"  : ([], []),
    "Merge Sort #16" : ([], []),
    "Merge Sort #32" : ([], []),
    "Merge Sort #64" : ([], []),
    "Merge Sort #128" : ([], []),
    "Merge Sort #256" : ([], []),
    "Merge Sort #512" : ([], []),
}

with open('sorts.txt') as csvfile:
    sortreader = csv.reader(csvfile, delimiter=',')
    
    for row in sortreader:
        sorts[row[0]][0].append(int(row[1]))
        sorts[row[0]][1].append(int(row[2]))
```

## Выводы

1. O(nlong) работает быстрее, чем O(n^2) при достаточно больших n. Не могу не отметить свое изумление тем фактом, что пузырьковая сортировка так сильно проигрывает остальным квадратичным.

2. Insertion sort с оптимизацией (перезаписывание, а не обмены) работает быстрее обычного так как совершает меньше действий и устраняет зависимости по данным, позволяя процессору эффективнее использовать конвейер. Еще сильнее эта идея развивается в Insertion sort с бинпоиском, где убрано чтение из массива при проверке условия цикла.

3. Оптимальный размер для оптимизации -- 64 элемента. Далее квадратичная природа Insertion Sort перевешивает рекурсивную природу Merge Sort.

4. Самая лучшая сортировка Radix Sort, так как она использует знание об элементах (что их можно поразрядно сравнивать)

5. qsort на случайном масиве работает во всех реализациях примерно одинаково, выделяется только реализаци со случайным опорным, так как она требует дорогого вызова rand()

6. qsort и лучший merge sort сравнимы по скорости

7. Как можно заметить по сравнению реализаций Qsort, на хороших массивах они работают примерно одинаково, в то время как на плохих данных стратегия выбора центрального ломается очень быстро, сводя ассимптотику к O(n²). Стратегия медианы работает лучше, однако тоже начинает проигрывать в несколько раз случайному выбору на неудачном массиве. Случайный выбор, хоть и несет дополнительные расходы на вызов rand(), всегда имеет ассимптотику O(nlogn) в силу малой вероятности (~10^(-13) для 20 элементов) [1]
